```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
```

```python
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({"scorecard" : "matchID"}, axis=1, inplace=True)
```

```python
df_match.head()
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]

    match_ids_dict[key1] = row["matchID"]
    match_ids_dict[key2] = row["matchID"]
```

Batting Summary

```python
with open("t20_json_files/t20_json_files/t20_wc_batting_summary.json") as f:
    data = json.load(f)

    records = []
    for rec in data:
        records.extend(rec['battingSummary'])
```

```python
df_batting = pd.DataFrame(records)
```

```python
df_batting.head()
```

```python
df_batting["decision"] = df_batting.dismissal.apply(lambda x: "out" if x else "not out")
```

```python
df_batting.head(10)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head()
```

```python
df_batting["batsmanName"] = df_batting.batsmanName.apply(lambda x: x.replace("â€", ""))
df_batting["batsmanName"] = df_batting.batsmanName.apply(lambda x: x.replace("\xa0", ""))
```

```python
df_batting.head(10)
```

```python
df_batting["matchID"] = df_batting["match"].map(match_ids_dict)
```

```python
df_batting.head()
```

```python
df_batting.to_csv("t20_csv_files/t20_csv_files/temp.csv", index=False)
```
